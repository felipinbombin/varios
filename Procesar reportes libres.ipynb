# Se define función para leer y filtrar reportes

```python
input_file_name = "input.csv"
input_file_path = os.path.join(project_path, input_file_name)

def read_input_file(start_date_threshold, end_date_threshold, route_filter=lambda x: True, stop_code_filter=lambda x: True):
    output = []

    with open(input_file_path, encoding='utf-8') as csvfile:
        spamreader = csv.reader(csvfile, delimiter='|', quotechar='"')
        next(spamreader)

        for row in spamreader:
            text = row[1].replace('\n', '')
            json_text = row[2]
            timestamp = row[3]
            phone_id = row[4]
            image_name = row[5]
            transappuser_id = row[6]
            image_file = row[7]

            timestamp = pytz.utc.localize(datetime.strptime(timestamp, '%Y-%m-%d %H:%M:%S.%f+00')).astimezone(tz)

            # filter by date
            if timestamp < start_date_threshold and timestamp > end_date_threshold:
                continue
            timestamp = timestamp.strftime("%Y-%m-%d %H:%M:%S")

            try:
                json_obj = json.loads(json_text)
            except json.JSONDecodeError:
                print(str(timestamp),"error con ", type(json_text), len(json_text))#, json_text)
                continue
            fields = []

            if 'bus' in json_obj:
                # asumo que es un bus
                route = json_obj['bus']['service']

                # filter by route
                if not route_filter(route):
                    continue

                fields.append(route)
                fields.append(json_obj['bus']['licensePlate'])
                fields.append(json_obj['locationUser']['latitude'].replace('/', ''))
                fields.append(json_obj['locationUser']['longitude'].replace('/', ''))
                fields.append(image_file)
            elif 'bus_stop' in json_obj:
                stop_code = json_obj['bus_stop']['id']
                
                if not stop_code_filter(stop_code):
                    continue
                
                fields.append(stop_code)
                fields.append("")
                fields.append(json_obj['locationUser']['latitude'].replace('/', ''))
                fields.append(json_obj['locationUser']['longitude'].replace('/', ''))
                fields.append(image_file)
            else:
                print(json_obj)

            output.append([text, timestamp, phone_id] + fields)

    return output
```

# Se define función para guardar resultado en CSV

```python
output_file_name = 'output.csv'
output_file_path = os.path.join(project_path, output_file_name)
    
def save_to_csv(row_list):
    header = ['\ufeffid', 'mensaje', 'fecha hora', 'teléfono', 'servicio o paradero', 'patente', 'latitud usuario', 'longitud usuario', 'imagen']

    with open(output_file_path, 'w', encoding='utf-8', newline='') as csvfilew:
        #csvfilew.write('\ufeff')
        spamwriter = csv.writer(csvfilew, delimiter=';', quotechar='"')
        spamwriter.writerow(header)

        for index, row in enumerate(row_list):
            spamwriter.writerow([index + 1] + row)
```
